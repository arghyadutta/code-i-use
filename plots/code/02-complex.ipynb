```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Note: If you use LaTeX, it will take slightly more time (~few seconds) to compile, but the math fonts looks nicer.

font_size = 10
matte_black = "#28282B"

font_settings = {
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "CMU Sans Serif",
    "font.size": font_size,
    "axes.labelsize": font_size,
    "legend.fontsize": font_size,
    "xtick.labelsize": font_size,
    "ytick.labelsize": font_size,
    "axes.titlesize": font_size,
}

axes_settings = {
    "axes.labelcolor": matte_black,
    "axes.edgecolor": matte_black,
    "axes.spines.top": False,
    "axes.spines.right": False,
    "xtick.color": matte_black,
    "ytick.color": matte_black,
    "xtick.direction": "out",
    "ytick.direction": "out",
    "axes.grid": True,
    "grid.alpha": 0.5,
    "grid.linestyle": ":",
    "grid.color": matte_black,
    "grid.linewidth": 0.5,
}

mpl.rcParams[
    "text.latex.preamble"
] = r"\usepackage[cm]{sfmath}\usepackage{amsmath}\usepackage{amssymb}"

mpl.rcParams.update(font_settings)
mpl.rcParams.update(axes_settings)
```

```python
x = np.arange(0, np.pi, 0.1)
sinx = np.sin(x)
cosx = np.cos(x)

# A single-column figure

fig, ax = plt.subplots(2, figsize=(3, 4), constrained_layout=True)


# A plot

color1 = "tab:red"
color2 = "tab:gray"

ax[0].plot(x, sinx, label="$\sin(x)$", color=color1)
ax[0].plot(x, cosx, label="$\cos(x)$", color=color2)

ax[0].set_xlabel("$x$")
ax[0].set_xticks([0, np.pi / 2, np.pi])
ax[0].set_xticklabels(["0", r"$\frac{\pi}{2}$", "$\pi$"])
ax[0].legend(handlelength=1, labelcolor="linecolor", frameon=False, loc="lower left")


# A histogram

# draw from normal distributions
mu = 1
sigma = 1

color3 = "tab:green"
color4 = "tab:blue"

ax[1].hist(
    np.random.normal(loc=mu, scale=sigma, size=10000),
    histtype="step",
    label="$\mathcal{N}(1,1)$",
    color=color3,
)
ax[1].hist(
    np.random.normal(loc=mu + 2, scale=sigma, size=10000),
    histtype="step",
    label="$\mathcal{N}(3,0.5)$",
    color=color4,
)

ax[1].set_ylabel("Count")
# Create a separate legend to avoid ugly boxes for histogram legends
custom_lines = [
    mpl.lines.Line2D([0], [0], color=color3),
    mpl.lines.Line2D([0], [0], color=color4),
]
ax[1].legend(
    custom_lines,
    ["$\mathcal{N}(1,1)$", "$\mathcal{N}(3,1)$"],
    handlelength=1,
    labelcolor="linecolor",
    frameon=False,
    handletextpad=0.2,
    loc="upper left",
    bbox_to_anchor=(0.7, 1),
)


# bbox_inches makes sure the texts at the sides of the figure are not cut while saving it as pdf
fig.savefig("../figures/complex.pdf", bbox_inches="tight");
```
