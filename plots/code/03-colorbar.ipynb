```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# plot heatmaps with seaborn and share colorbars
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html

A = np.random.uniform(low=-20, high=-10, size=(50, 50))
B = np.random.uniform(low=0, high=10, size=(50, 50))
C = np.random.uniform(low=20, high=30, size=(50, 50))


fig, ax = plt.subplots(1, 3, figsize=(5, 2), constrained_layout=True)

# add axis for common colorbar. it adds a rectangular patch at [left, bottom, width, height]
cax = fig.add_axes([1, 0.1, 0.01, 0.8])

# the important thing while sharing colorbars among subplots is to set both cbar_ax and vmin,vmax. See the second answer in [here](https://stackoverflow.com/questions/28356359/one-colorbar-for-seaborn-heatmaps-in-subplot)

_vmin = min([A.min(), B.min(), C.min()])
_vmax = max([A.max(), B.max(), C.max()])
# print(_vmin, _vmax)

heatmap_prop = dict(
    vmin=_vmin,
    vmax=_vmax,
    cbar_ax=cax,
    cbar_kws=dict(label="color_bar_label"),
)
sns.heatmap(A, ax=ax[0], **heatmap_prop)
sns.heatmap(B, ax=ax[1], **heatmap_prop)
sns.heatmap(C, ax=ax[2], **heatmap_prop)

# set colorbar ticklabels depending on the range of values.
cax.set_yticks(np.arange(-20, 40, 10))

ax[0].set_title("A")
ax[1].set_title("B")
ax[2].set_title("C")

pass;
```
